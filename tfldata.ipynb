```python
import urllib.request
import json

try:
    url = "https://api.tfl.gov.uk/Journey/JourneyResults/SW19%202NZ/to/EC3M%208AF?timeIs=Arriving&journeyPreference=LeastInterchange&accessibilityPreference=NoRequirements&walkingSpeed=Slow&cyclePreference=None&bikeProficiency=Easy"
    hdr = {'Cache-Control': 'no-cache'}

    req = urllib.request.Request(url, headers=hdr)
    response = urllib.request.urlopen(req)

    # Read the response
    response_data = response.read()

    # Parse the response data as JSON
    data = json.loads(response_data)

    # Now you can work with 'data' as a Python dictionary
    print(json.dumps(data, indent=2))  # This is for pretty printing the JSON data

except Exception as e:
    print(e)
```

```python
# Assuming 'data' is your parsed JSON response from the TfL API
journeys = data['journeys']
```

```python
# Check the number of journeys
print(f"Number of journeys: {len(journeys)}")

# Optionally, inspect the first journey
if journeys:  # Check if the list is not empty
    print(journeys[0])  # Print the first journey to inspect its structure
```

```python
import folium

def add_leg_to_map(_map, leg):
    # Extract start and end coordinates
    start_coord = (leg['departurePoint']['lat'], leg['departurePoint']['lon'])
    end_coord = (leg['arrivalPoint']['lat'], leg['arrivalPoint']['lon'])

    # Add markers
    folium.Marker(start_coord, popup='Start', icon=folium.Icon(color='green')).add_to(_map)
    folium.Marker(end_coord, popup='End', icon=folium.Icon(color='red')).add_to(_map)

    # Draw the leg path if available
    if 'path' in leg and 'lineString' in leg['path']:
        path_coords = [[float(coord) for coord in point.split(',')] for point in leg['path']['lineString'][2:-2].split('],[')]
        folium.PolyLine(locations=path_coords, color='blue').add_to(_map)

# Initialize the base map
first_leg = journeys[0]['legs'][0]
base_map = folium.Map(location=[first_leg['departurePoint']['lat'], first_leg['departurePoint']['lon']], zoom_start=14)

# Add each leg to the map
for journey in journeys:
    for leg in journey['legs']:
        add_leg_to_map(base_map, leg)

# Display the map
base_map
```
